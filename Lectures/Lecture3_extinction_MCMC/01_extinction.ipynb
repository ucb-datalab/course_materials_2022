* Extinction is the absorption and scattering of electromagnetic radiation by dust and gas between an emitting astronomical object and the observer.

* The effects of extinction were first documented by Berkeley Astronomer Robert Trumpler in 1930.

 ### Astrophysics of Extinction
 
 <img src="http://english.shao.cas.cn/ns/es/201503/W020150323356387826647.jpg"></img>
 
 * Consider a hot object (e.g., a star) emitted light with a specific flux $F_{\lambda, 0}$ that is some distance $d$ away from us.

 * Along the line of sight there is some dust.  The dust has some (constant) number density $n$ and each dust grain has a cross-section $\sigma_{\lambda}$.
 
 
 * We can define the quanity optical depth, $\tau_{\lambda} = \int_0^d \sigma_{\lambda} n ds$.  $\tau$ is dimesionless and effectively measures the average number of times a photon scatters as it travels through a medium. 
 
  - $\tau >> 1$ is called __optically thick__ which means a photon has a short mean free path (the average distance it travels before scattering). 
  
  - $\tau << 1$ is called __optically thin__ which means a photon has a long mean free path.
  
  
 <img src="optical_thickness.jpg" width=80%></img> 
  
 
 * Consider a $F_{\lambda}$ is the specific flux we receive at Earth from a hot source, moduldated by the light that is scattered out of the line-of-sight by the intervening dust.  This can be expressed mathematically as:
 
    $F_{\lambda} = F_{\lambda, 0} e^{-\tau_{\lambda}} = F_{\lambda, 0} e^{- \int_0^d \sigma_{\lambda} n ds} = F_{\lambda, 0} e^{- \sigma_{\lambda} n d}$ 
 
 where the exponential term comes from solving the equation of radiative transfer.


* The ratio of observed to incident flux is then:

    $\frac{F_{\lambda}}{F_{\lambda, 0}} = e^{-\tau_{\lambda}}$
    
* and can be written in terms of magnitudes as:    
    
    $-2.5\log_{10}(\frac{F_{\lambda}}{F_{\lambda, 0}}) = -2.5\log_{10}(e^{-\tau_{\lambda}}) \rightarrow$
    
    $m_{\lambda} \sim 1.086 \tau_{\lambda} = A_{\lambda}$.
    
* $A_{\lambda}$ is called __selective absoprtion__.  It is measured in practice by comparing the spectra from two stars of same spectral type, one with no/low extinction.
 
* __Color Excess__ is the measured color of an object minus its intrinsic color:

    $E_{\lambda  1 - \lambda 2} = (m_{\lambda  1} -m_{\lambda 2}) - (m_{\lambda  1} -m_{\lambda 2})_0$
    

* In practice, for historical reasons, we often use $A_V$ and $E(B-V)$ (i.e., blue and visual filters)

* The relation between selective absoption and color excess is given by

    $R_V = \frac{A_V}{E(B-V)}$
    
* In the Milky Way, $R_V = 3.1$ (more or less)
